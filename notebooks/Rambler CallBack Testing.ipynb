```python
import deep_book as dp
```

```python
nlp, doc = dp.nlp_doc()
```

```python
indexer = dp.Indexer(doc)
```

```python
d = indexer.generator()
```

```python
m = dp.ai.lstm(indexer)
```

```python
rambler = dp.Rambler(indexer=indexer, model=m, nlp=nlp)
rambler(u'')
```

```python
m.summary()
```

```python
m.compile("rmsprop", loss = "sparse_categorical_crossentropy")
```

```python
rambler_cb = dp.RamblerCallback(rambler)
```

```python
m.fit_generator(d, steps_per_epoch=10, epochs=5, callbacks=[rambler_cb])
```
